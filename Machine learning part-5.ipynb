Theory question


1. What is a Decision Tree, and how does it workA Decision Tree is a supervised machine learning algorithm used for both classification and regression tasks. It works by splitting the dataset into subsets based on the value of input features. This process is repeated recursively, creating a tree-like structure with decision nodes and leaf nodes representing outcomes.

2. What are impurity measures in Decision TreesImpurity measures evaluate the quality of a split. Common impurity measures include Gini Impurity and Entropy. They quantify how mixed the classes are in the nodes.

3. What is the mathematical formula for Gini ImpurityGini(t) = 1 - ∑ p(i)^2where p(i) is the probability of an element being classified into a particular class.

4. What is the mathematical formula for EntropyEntropy(t) = - ∑ p(i) * log2(p(i))where p(i) is the probability of class i in the node.

5. What is Information Gain, and how is it used in Decision TreesInformation Gain is the reduction in entropy after a dataset is split on an attribute. It helps in selecting the feature that best separates the data.

6. What is the difference between Gini Impurity and EntropyGini is computationally less intensive and performs well. Entropy is more information-theoretic. Both give similar results but may choose different splits.

7. What is the mathematical explanation behind Decision TreesThey use recursive binary splitting based on impurity measures to divide data. At each node, the feature and threshold that yield the best split (lowest impurity) are chosen.

8. What is Pre-Pruning in Decision TreesPre-Pruning involves stopping the tree growth early based on criteria like max_depth, min_samples_split, etc., to avoid overfitting.

9. What is Post-Pruning in Decision TreesPost-Pruning grows a full tree and then removes branches that do not contribute significantly to model accuracy, improving generalization.

10. What is the difference between Pre-Pruning and Post-PruningPre-pruning halts tree construction early. Post-pruning removes parts of a fully grown tree. Post-pruning often provides better performance.

11. What is a Decision Tree RegressorA Decision Tree Regressor predicts continuous values by splitting the dataset into regions with similar output values and averaging them.

12. What are the advantages and disadvantages of Decision TreesAdvantages: Easy to interpret, requires little data preprocessing.Disadvantages: Prone to overfitting, unstable with small data changes.

13. How does a Decision Tree handle missing valuesTechniques include surrogate splits, missing value imputation, or branching both ways during training.

14. How does a Decision Tree handle categorical featuresThey can split on each category or use one-hot encoding before training.

15. What are some real-world applications of Decision Trees?Medical diagnosis, credit scoring, fraud detection, customer churn prediction, etc.

```python
# Practical Questions on Decision Trees (with question numbers)

# 1. Write a Python program to train a Decision Tree Classifier on the Iris dataset and print the model accuracy
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=42)
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Q1 Accuracy:", accuracy_score(y_test, y_pred))

# 2. Write a Python program to train a Decision Tree Classifier using Gini Impurity and print feature importances
clf_gini = DecisionTreeClassifier(criterion='gini')
clf_gini.fit(X_train, y_train)
print("\nQ2 Feature Importances (Gini):", clf_gini.feature_importances_)

# 3. Write a Python program to train a Decision Tree Classifier using Entropy and print the model accuracy
clf_entropy = DecisionTreeClassifier(criterion='entropy')
clf_entropy.fit(X_train, y_train)
y_pred_entropy = clf_entropy.predict(X_test)
print("\nQ3 Accuracy (Entropy):", accuracy_score(y_test, y_pred_entropy))

# 4. Write a Python program to train a Decision Tree Regressor on housing dataset and evaluate using MSE
from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

housing = fetch_california_housing()
X_train_h, X_test_h, y_train_h, y_test_h = train_test_split(housing.data, housing.target, test_size=0.3, random_state=42)
reg = DecisionTreeRegressor()
reg.fit(X_train_h, y_train_h)
y_pred_h = reg.predict(X_test_h)
print("\nQ4 MSE:", mean_squared_error(y_test_h, y_pred_h))

# 5. Visualize a Decision Tree using graphviz
from sklearn.tree import export_graphviz
import graphviz

dot_data = export_graphviz(clf, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
graph = graphviz.Source(dot_data)
graph.render("Q5_decision_tree")  # Saves as Q5_decision_tree.pdf

# 6. Decision Tree Classifier with max depth 3 and compare accuracy
clf_depth3 = DecisionTreeClassifier(max_depth=3)
clf_depth3.fit(X_train, y_train)
y_pred_depth3 = clf_depth3.predict(X_test)
print("\nQ6 Accuracy (depth=3):", accuracy_score(y_test, y_pred_depth3))

# 7. Decision Tree with min_samples_split=5
clf_min_split = DecisionTreeClassifier(min_samples_split=5)
clf_min_split.fit(X_train, y_train)
y_pred_min_split = clf_min_split.predict(X_test)
print("\nQ7 Accuracy (min_samples_split=5):", accuracy_score(y_test, y_pred_min_split))

# 8. Feature scaling before training (optional, tree-based models don't need it, but for comparison)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
clf_scaled = DecisionTreeClassifier()
clf_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = clf_scaled.predict(X_test_scaled)
print("\nQ8 Accuracy (with scaling):", accuracy_score(y_test, y_pred_scaled))

# 9. One-vs-Rest strategy (default for DecisionTreeClassifier)
# Already handled by default DecisionTreeClassifier in scikit-learn. Reusing clf from Q1.
print("\nQ9 OvR is default strategy in DecisionTreeClassifier")

# 10. Display feature importances
print("\nQ10 Feature Importances:", clf.feature_importances_)

# 11. Decision Tree Regressor with max_depth=5 vs unrestricted
reg_depth5 = DecisionTreeRegressor(max_depth=5)
reg_depth5.fit(X_train_h, y_train_h)
y_pred_depth5 = reg_depth5.predict(X_test_h)
print("\nQ11 MSE (max_depth=5):", mean_squared_error(y_test_h, y_pred_depth5))

# 12. Cost Complexity Pruning
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas
accuracies = []

for alpha in ccp_alphas:
    clf_pruned = DecisionTreeClassifier(ccp_alpha=alpha)
    clf_pruned.fit(X_train, y_train)
    acc = accuracy_score(y_test, clf_pruned.predict(X_test))
    accuracies.append(acc)
print("\nQ12 Accuracies with CCP:", accuracies)

# 13. Evaluate with Precision, Recall, and F1-Score
from sklearn.metrics import classification_report
print("\nQ13 Classification Report:\n", classification_report(y_test, y_pred))

# 14. Confusion matrix visualization
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.title("Q14 Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# 15. Use GridSearchCV to tune max_depth and min_samples_split
from sklearn.model_selection import GridSearchCV

param_grid = {'max_depth': [2, 3, 4, None], 'min_samples_split': [2, 5, 10]}
grid = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5)
grid.fit(X_train, y_train)
print("\nQ15 Best Parameters:", grid.best_params_)
print("Q15 Best Accuracy:", grid.best_score_)
```
